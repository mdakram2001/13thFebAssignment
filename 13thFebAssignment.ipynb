```python
"""Q4: Why LookupError class is used? Explain with an example KeyError and IndexError.
Ans:

The LookupError class in Python is used to define errors that occur when a specified key or index cannot be found
in a collection."""

#KeyError: This error is raised when a specified key cannot be found in a dictionary.

d = {'a': 1, 'b': 2, 'c': 3}

try:
    value = d['d']
except KeyError as e:
    print(e)

#IndexError: This error is raised when a specified index is out of range of a sequence.

lst = [1, 2, 3]

try:
    value = lst[3]
except IndexError as e:
    print(e)
```
